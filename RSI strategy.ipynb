#### Defining RSI

```python
def RSIcalc(asset): # the asset here will be the ticker symbol
    df=yf.download(asset, start="2011-01-01", end="2020-12-31")
    df["ma_200"]=df["Adj Close"].rolling(200).mean() # Calculating the 200 day MA
    df["delta"]=df["Adj Close"]. pct_change() # Just simple daily returns
    df["upmove"]=df["delta"].apply(lambda x:x if x>0 else 0) # Definition of Relative strength
    df["downmove"]=df["delta"].apply(lambda x:abs(x) if x<0 else 0) # Note that the lambda function is basically applying the 
                                                                    # given condition to every single row in the delta column
    df["avg_up"]=df["upmove"].ewm(span=19).mean()
    df["avg_down"]=df["downmove"].ewm(span=19).mean()
    
    # we are rolling over 200 days. This means the first 200 days will be completely filled with NaN values. Hence, now we gotta
    # drop them
    
    df=df.dropna()
    df["rs"]=df["avg_up"]/df["avg_down"]
    df["rsi"]=df["rs"].apply(lambda x:100-(100/(x+1)))
    
    # defining the conditions when to buy and sell assets
    
    df.loc[(df["Adj Close"]>df["ma_200"]) & (df["rsi"]< 30), "Buy"]="Yes"
    df.loc[(df["Adj Close"]< df["ma_200"]) | (df["rsi"] > 30), "Buy"]="No"
        
    return df
    
```

```python
# This function is basically taking the df as an argument and will provide us the buying and selling dates and This will 
# ultimately help us to calculate the profits of this rsi strategy

buying_dates=[]
selling_dates=[]

def getSignals(df): 
    for i in range(len(df)-11):
        if "Yes" in df["Buy"].iloc[i]: # Checking if there is "Yes" in the Buy column 
            buying_dates.append(df.iloc[i+1].name) # The ".name" here means we are taking the timestamp of the row(i.e. date)
            for j in range(1,11): # for holding period of 10 days, range 1 to 11 coz 11 is excluded
                if df["rsi"].iloc[i+j] > 40:
                    selling_dates.append(df.iloc[i+j+1].name)
                    break
                elif j==10:
                    selling_dates.append(df.iloc[i+j+1].name)

    return buying_dates, selling_dates
    
```

```python
frame=RSIcalc(tickers[0])
buy,sell= getSignals(frame)
```

```python
# visualizing the above buying signals

plt.figure(figsize= (12,5))
plt.scatter(frame.loc[buy].index, frame.loc[buy]["Adj Close"], marker="^", c="g")
plt.plot(frame["Adj Close"], alpha=0.7)
```

```python
profits=((frame.loc[sell].Open.values) / (frame.loc[buy].Open.values))-1
```

```python
profits
```

```python
# Calculating the winning rate

wins= [i for i in profits if i > 0]
```

```python
len(wins)/len(profits) 

# Hence, we are getting a win rate of about 84 % for the ticker 3M
```

#### Now, lets apply this to all the stocks in the s & p 500

```python
matrix_signals=[] # Here we are storing our buying signals from all the stocks
matrix_profits=[]

for i in range(len(tickers)):
    frame=RSIcalc(tickers[i])
    buy,sell= getSignals(frame)
    profits=(frame.loc[sell].Open.values/frame.loc[buy].Open.values)-1
    matrix_signals.append(buy)
    matrix_profits.append(profits)
    
```
